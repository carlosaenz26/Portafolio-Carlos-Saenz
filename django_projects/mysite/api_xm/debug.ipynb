```python
# Bloque 1: Configuración y Librerías
import pandas as pd
from pydataxm import *
import logging
import plotly.express as px
from datetime import datetime, timedelta
import nbformat  # Añadido para verificar

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

# Verificar nbformat

# Bloque 2: Conexión y Colecciones
logger.info("Conectando a la API de XM...")
objetoAPI = pydataxm.ReadDB()
variables_xm = objetoAPI.get_collections()

if variables_xm is None or variables_xm.empty:
    logger.warning("No se obtuvieron colecciones")
else:
    logger.info(f"Total de métricas obtenidas: {len(variables_xm)}")
    display(variables_xm.head())
    logger.info(f"Columnas disponibles: {list(variables_xm.columns)}")

# Bloque 3: Análisis de Estructura
logger.info("Explorando tipos de datos y valores únicos...")
for col in variables_xm.columns:
    logger.info(f"{col}: {variables_xm[col].dtype}, Valores únicos: {variables_xm[col].nunique()}")

unique_types = variables_xm['Type'].unique()
logger.info(f"Tipos únicos: {unique_types}")
unique_entities = variables_xm['Entity'].unique()
logger.info(f"Entidades únicas: {unique_entities}")

# Bloque 4: Guardar Referencia
variables_xm.to_csv("variables_xm.csv", index=False)
logger.info("Datos guardados en variables_xm.csv")

# Bloque 5: Consultar Datos de Ejemplo
logger.info("Consultando datos de ejemplo para 'DemaReal' (Sistema)...")
metric_id = "DemaReal"
entity = "Sistema"
start_date = "2024-01-01"  # Asegúrate de que estas fechas tengan datos disponibles
end_date = "2024-01-07"
data_df = objetoAPI.request_data(metric_id, entity, start_date, end_date)

if data_df is None or data_df.empty:
    logger.warning(f"No se obtuvieron datos para {metric_id} ({entity})")
else:
    logger.info(f"Datos obtenidos para {metric_id} ({entity}): {len(data_df)} filas")
    display(data_df.head())
    logger.info(f"Columnas de datos: {list(data_df.columns)}")

# Bloque 6: Transformar Datos Horarios a Formato Largo
if data_df is not None and not data_df.empty:
    # Identificar columnas de valores horarios
    hour_cols = [col for col in data_df.columns if col.startswith("Values_Hour")]
    logger.info(f"Columnas horarias encontradas: {hour_cols}")

    # Convertir a formato largo
    data_long = pd.melt(
        data_df,
        id_vars=["Date", "Id", "Values_code"],  # Columnas fijas
        value_vars=hour_cols,                   # Columnas a transformar
        var_name="Hour",                        # Nombre de la nueva columna de horas
        value_name="Value"                      # Nombre de la nueva columna de valores
    )

    # Crear una columna combinada de fecha y hora
    data_long["Hour"] = data_long["Hour"].str.replace("Values_Hour", "").astype(int) - 1  # Convertir a 0-23
    data_long["Date_Hour"] = pd.to_datetime(data_long["Date"]) + pd.to_timedelta(data_long["Hour"], unit="h")
    logger.info("Datos transformados a formato largo:")
    display(data_long.head())

# Bloque 7: Visualización (Modo Offline)
if data_long is not None and not data_long.empty:
    fig = px.line(data_long, x="Date_Hour", y="Value", title=f"{metric_id} ({entity}) - {start_date} a {end_date}")
    fig.update_layout(xaxis_title="Fecha y Hora", yaxis_title="Valor (kWh)")
    fig.write_html("plot.html")  # Guarda como archivo HTML
    logger.info("Gráfica guardada como plot.html; ábrelo en un navegador")
else:
    logger.warning("No hay datos transformados para visualizar")
```
